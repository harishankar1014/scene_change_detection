```python
def grabcut(img):
    mask = np.zeros(img.shape[:2],np.uint8)#GC_BGD,GC_FGD,GC_PR_BGD,GC_PR_FGD 
    bgdModel = np.zeros((1,65),np.float64)
    fgdModel = np.zeros((1,65),np.float64)
#     rect = (0,0,300,300)
    rect = (0,0,260,160)
    cv.grabCut(img,mask,rect,bgdModel,fgdModel,5,cv.GC_INIT_WITH_RECT)
    mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
    img_masked = img*mask2[:,:,np.newaxis]
    return img_masked
```

```python
# im1 = img_array[0]
# im2 = img_array[10]

im1_gray = cv.cvtColor(im1,cv.COLOR_BGR2GRAY)
im2_gray = cv.cvtColor(im2,cv.COLOR_BGR2GRAY)

sz = im1.shape

# warp_mode = cv.MOTION_HOMOGRAPHY
warp_mode = cv.MOTION_EUCLIDEAN

if warp_mode == cv.MOTION_HOMOGRAPHY :
    warp_matrix = np.eye(3, 3, dtype=np.float32)
else:
    warp_matrix = np.eye(2, 3, dtype=np.float32)
print("Wrap mode :",warp_mode) 
number_of_iterations = 5;

termination_eps = 1e-10;

criteria = (cv.TERM_CRITERIA_EPS | cv.TERM_CRITERIA_COUNT, number_of_iterations,  termination_eps)

(cc, warp_matrix) = cv.findTransformECC (im1_gray,im2_gray,warp_matrix, warp_mode, criteria)
print("Completed find transform")
if warp_mode == cv.MOTION_HOMOGRAPHY :
    im2_aligned = cv.warpPerspective(im2, warp_matrix, (sz[1],sz[0]), flags=cv.INTER_LINEAR + cv.WARP_INVERSE_MAP)
else:
    im2_aligned = cv.warpAffine(im2, warp_matrix, (sz[1],sz[0]), flags=cv.INTER_LINEAR + cv.WARP_INVERSE_MAP)

print("Correlation coefficient:",cc)    
```

```python
row = im1.shape[0]
col = im1.shape[1]

tiles1 = []
tiles2 = []
mapping = []
counter = 0
block_size = 32
for x in range(0, row, block_size):
    r = []
    for y in range(0, col, block_size):
        tiles1.append(im1_gray[x:x+block_size,y:y+block_size])
        tiles2.append(im2_aligned_gray[x:x+block_size,y:y+block_size])
        counter = counter + 1
        r.append(counter)
    mapping.append(r)
```

```python
import pywt
dwt_img1 = []
dwt_aligned = []
for i in range(len(tiles1)):
    (cA,(cH,cV,cD)) = pywt.dwt2(tiles1[i], 'haar')
    dwt_img1.append(np.array(cA))
    (cA1,(cH1,cV1,cD1)) = pywt.dwt2(tiles2[i], 'haar')
    dwt_aligned.append(np.array(cA1))
```

```python
def pcc_single(cA,cR):
    pcc = np.corrcoef(cA.flatten(),cR.flatten())[0,1]
    return pcc
```

```python
pcc_result_dwt = []
for i in range(len(dwt_aligned)):
    result = pcc_single(dwt_img1[i],dwt_aligned[i])
    pcc_result_dwt.append(result)
    
# print("This is the pcc result for dwt")
# pcc_result_dwt[0]
```

```python
#Gabor filters
def build_filters(_lambda,_orientation):
    filters = []
    ksize = 31
    theta = np.radians(_orientation)
    kern = cv.getGaborKernel((ksize, ksize), 4.0, theta, _lambda, 0.5, 0, ktype=cv.CV_32F)
    kern /= 1.5*kern.sum()
    filters.append(kern)
    return filters

def process(img, filters):
    accum = np.zeros_like(img)
    for kern in filters:
        fimg = cv.filter2D(img, cv.CV_8UC3, kern)
        np.maximum(accum, fimg, accum)
    return accum

filters0 = build_filters(20,0)
filters10 = build_filters(80,10)
filters20 = build_filters(30,20)
filters30 = build_filters(10,30)
filters40 = build_filters(20,40)
filters50 = build_filters(80,50)
filters60 = build_filters(30,60)
filters70 = build_filters(10,70)
filters80 = build_filters(20,80)
filters90 = build_filters(80,90)
filters100 = build_filters(30,100)
filters110 = build_filters(10,110)
filters120 = build_filters(20,120)
filters130 = build_filters(80,130)
filters140 = build_filters(30,140)
filters150 = build_filters(10,150)
filters160 = build_filters(20,160)
filters170 = build_filters(80,170)
filters180 = build_filters(30,180)
```

```python
def ACO_saliencymap(gray_image):
    filtered_responses = []
    
    filtered_responses.append(process(gray_image, filters0))
    filtered_responses.append(process(gray_image, filters10))
    filtered_responses.append(process(gray_image, filters20))
    filtered_responses.append(process(gray_image, filters30))
    filtered_responses.append(process(gray_image, filters40))
    filtered_responses.append(process(gray_image, filters50))
    filtered_responses.append(process(gray_image, filters60))
    filtered_responses.append(process(gray_image, filters70))
    filtered_responses.append(process(gray_image, filters80))
    filtered_responses.append(process(gray_image, filters90))
    filtered_responses.append(process(gray_image, filters100))
    filtered_responses.append(process(gray_image, filters110))
    filtered_responses.append(process(gray_image, filters120))
    filtered_responses.append(process(gray_image, filters130))
    filtered_responses.append(process(gray_image, filters140))
    filtered_responses.append(process(gray_image, filters150))
    filtered_responses.append(process(gray_image, filters160))
    filtered_responses.append(process(gray_image, filters170))
    filtered_responses.append(process(gray_image, filters180))
    
    feature_maps = np.abs(filtered_responses)
    saliency_map = np.sum(feature_maps, axis=0)
#     print(gray_image.shape)
#     print(saliency_map.shape)
    phermones = np.ones_like(saliency_map)
    evaporation_rate = 0.5
    iterations = 100
    
    for _ in range(iterations):
        normalized_phermone = phermones/np.max(phermones)
        updated_saliency_map = saliency_map * normalized_phermone
    
        local_contrast = cv.Laplacian(updated_saliency_map,cv.CV_64F)
        local_contrast = np.abs(local_contrast)
        phermone = (1-evaporation_rate)*phermones + local_contrast
        
    final_saliency_map = cv.normalize(updated_saliency_map.astype(np.float32),None,alpha=0,beta=255,norm_type=cv.NORM_MINMAX)
    final_saliency_map = cv.GaussianBlur(final_saliency_map.astype(np.float32),(5,5),0)
    return final_saliency_map
```

```python
def draw_grid(img):
    line_color=(255, 0, 0)
    thickness=1
    type_=cv.LINE_AA
    pxstep=32
    x = pxstep
    y = pxstep
    while x < img.shape[1]:
        cv.line(img, (x, 0), (x, img.shape[0]), color=line_color, lineType=type_, thickness=thickness)
        x += pxstep

    while y < img.shape[0]:
        cv.line(img, (0, y), (img.shape[1], y), color=line_color, lineType=type_, thickness=thickness)
        y += pxstep
        
```
